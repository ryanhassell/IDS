

For updating the NULL value in the SQL table, I used this SQL command to replace the value with the average:

```
UPDATE USArrests
SET Assault = (
    SELECT AVG(Assault)
    FROM USArrests
    WHERE Assault is NOT NULL
) WHERE "Georgia" IS State;
```



For the min, max, mean, and variance of the table, I used these commands:

```
SELECT
    MIN("Murder") AS min,
    MAX("Murder") AS max,
    AVG("Murder") AS mean,
    VARIANCE("Murder") AS variance
FROM
    USArrests
```
```
SELECT
    MIN("Assault") AS min,
    MAX("Assault") AS max,
    AVG("Assault") AS mean,
    VARIANCE("Assault") AS variance
FROM
    USArrests
```
```
SELECT
    MIN("UrbanPop") AS min,
    MAX("UrbanPop") AS max,
    AVG("UrbanPop") AS mean,
    VARIANCE("UrbanPop") AS variance
FROM
    USArrests
```



In order to fix and replace data with medians, I used these commands:


```
UPDATE childmortality
SET "Underve_mortality_rate" = (SELECT MEDIAN("Underve_mortality_rate") FROM childmortality)
WHERE "Underve_mortality_rate" IS NULL;

UPDATE childmortality
SET "Infant_mortality_rate" = (SELECT MEDIAN("Infant_mortality_rate") FROM childmortality)
WHERE "Infant_mortality_rate" IS NULL;

UPDATE childmortality
SET "Neonatal_mortality_rate" = (SELECT MEDIAN("Neonatal_mortality_rate") FROM childmortality)
WHERE "Neonatal_mortality_rate" IS NULL;
```



And to get the years with the most and least infant deaths, I used the commands:


```
SELECT Year as "Year with Minimum"
FROM childmortality
WHERE infant_mortality_rate = (SELECT MIN(infant_mortality_rate) FROM Childmortality);

SELECT Year as "Year with Maximum"
FROM childmortality
WHERE infant_mortality_rate = (SELECT MAX(infant_mortality_rate) FROM Childmortality);
```
